```python
import requests # Traemos la librerías que necesitamos
import pandas as pd 

# Hacemos la petición a la API.
response = requests.get('https://beta.adalab.es/resources/apis/pelis/pelis.json') # Esta dirección es el famoso "endpoint".

# Verificamos si la petición fue exitosa (código 200)
if response.status_code == 200:
    # Mostramos los datos del API
    print(response.json())
    data = response.json()
else:
    # Mostramos un mensaje si hubo un error
    print(f"Error en la petición. Código de estado: {response.status_code}")

```

Fase 3: Inserción de los Datos en la Base de Datos

```python
mycursor = cnx.cursor() 
# abrimos el cursor asociado a esta conexión
```

```python
# CREACION TABLA PELICULAS API

tabla_peliculas_API = """
                        CREATE TABLE IF NOT EXISTS peliculas_API (
                            id INT AUTO_INCREMENT PRIMARY KEY,
                            titulo VARCHAR(100) NOT NULL,
                            año YEAR NOT NULL,
                            duracion INT NOT NULL,
                            genero VARCHAR(50) NOT NULL,
                            adultos BOOLEAN,
                            sub_es BOOLEAN DEFAULT FALSE,
                            sub_en BOOLEAN DEFAULT FALSE,
                            sub_fr BOOLEAN DEFAULT FALSE
                        );
                        """
                    

mycursor.execute(tabla_peliculas_API)
cnx.commit()
print("Tabla creada correctamente")
```

```python
for pelicula in data[:100]:  # limitamos a 100 películas
    # Revisamos qué subtítulos tiene la película
    subs = pelicula.get("subtitulos", [])
    sub_es = "es" in subs
    sub_en = "en" in subs
    sub_fr = "fr" in subs

    sql = """
        INSERT INTO peliculas_API (titulo, año, duracion, genero, adultos, sub_es, sub_en, sub_fr)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """
    valores = (
        pelicula.get("titulo"),
        pelicula.get("año"),
        pelicula.get("duracion"),
        pelicula.get("genero"),
        pelicula.get("adultos"),
        sub_es,
        sub_en,
        sub_fr
    )

    mycursor.execute(sql, valores)


# Guardar los cambios
cnx.commit()
print("¡Datos insertados correctamente!")
cnx.close() # Cierra la conexión.
print("Conexión cerrada")
```

2. ¿Cuántas películas incluyen subtítulos en español?

4. ¿Cuál es la película más antigua registrada en la base de datos?

```python
mycursor.execute("""
    SELECT titulo, año
    FROM peliculas_api
    ORDER BY año ASC
    LIMIT 1
   
""")

# Obtenemos el resultado

pelicula_antigua = mycursor.fetchone()
print(f"La película más antigua es '{pelicula_antigua[0]}' del año {pelicula_antigua[1]}")
```

5. Muestra el promedio de duración de las películas agrupado por género.

```python
mycursor.execute("""
    SELECT genero, AVG(duracion) AS promedio_duracion
    FROM peliculas_api
    GROUP BY genero
   
""")

# Obtenemos el resultado

resultado = mycursor.fetchall()
# Mostramos los resultados
for genero, promedio in resultado:
    
    print(f"Género: {genero}, Duración promedio: {promedio} minutos")
```

6. ¿Cuántas películas por año se han registrado en la base de datos? Ordena de mayor a menor.

```python
mycursor.execute("""
    SELECT año, COUNT(*) AS cantidad_peliculas
    FROM peliculas_api
    GROUP BY año
    ORDER BY cantidad_peliculas DESC
   
""")

# Obtenemos el resultado

resultado = mycursor.fetchall()
# Mostramos los resultados
for año,cantidad_peliculas in resultado:
    
     print(f"Año: {año}, Peliculas registradas: {cantidad_peliculas}")
```

7. ¿Cuál es el año con más películas en la base de datos.

```python
mycursor.execute("""
    SELECT año, COUNT(*) AS peliculas_registradas
    FROM peliculas_api
    GROUP BY año
    ORDER BY peliculas_registradas DESC
    LIMIT 1
   
""")

# Obtenemos el resultado

resultado = mycursor.fetchone()
# Mostramos los resultados
año,peliculas_registradas = resultado
    
print(f"El año con más peliculas registradas es:{año} con {peliculas_registradas} peliculas")
```

8. Obtén un listado de todos los géneros y cuántas películas corresponden a cada uno.

```python
mycursor.execute("""
    SELECT genero,COUNT(titulo) AS total
    FROM peliculas_api
    GROUP BY genero
     
 
""")

# Obtenemos el resultado

resultado = mycursor.fetchall()
# Mostramos los resultados

for genero, total in resultado:   
    print(f"Genero{genero}: cantidad {total} peliculas")
```
